```python
import math
import random
def calcLog(x, base):
    base = 10
    n = 0
    while x > base:
        x = x / base
        n += 1
    return(n)

antall_testverdier = 1000
max_testverdi = 1e9

for i in range(antall_testverdier):
    x = random.randint(1,max_testverdi)
    min_log = calcLog(x,10)
    python_log = math.log10(x)
    if (round(python_log) != min_log and round(python_log) != min_log+1):
        print(f"Gjetning: log_{base} av {x} er mellom {min_log} og {min_log+1}. Nøyaktig: {python_log}")
```
